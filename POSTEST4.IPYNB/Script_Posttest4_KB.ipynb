#### 3.) DATA CLEANING

#### LANGKAH PERTAMA MEMBUKA FILE CSV YANG AKAN DIGUNAKAN

```python
df = pd.read_csv("Employee.csv")
```

#### CEK TOTAL NILAI NULL PADA DATASET

```python
df.isnull().sum()
```

#### CEK LETAK NILAI NULL PADA DATASET

```python
df[df.isnull().any(axis = 1)]
```

#### MENANGANI NILAI NULL DENGAN PANDAS

#### BILANGAN BULAT DENGAN MENGGUNAKAN MEDIAN

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df.isna().sum()
```

#### TULISAN DENGAN MENGGUNAKAN MODUS

```python
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
```

```python
df.isna().sum()
```

#### BILANGAN DESIMAL DENGAN MENGGUNAKAN MEAN

```python
df['ExperienceInCurrentDomain'] = df['ExperienceInCurrentDomain'].fillna(df['ExperienceInCurrentDomain'].mean())
```

```python
df.isna().sum()
```

#### 4.) MENGUBAH TIPE DATA SALAH SATU ATTRIBUT ANGKA

#### CEK TIPE DATA PADA DATASET

```python
df.info()
```

#### KITA AKAN MENCOBA MENGUBAH 'LEAVEORNOT' YANG BERTIPE INTEGER MENJADI OBJECT

```python
df['LeaveOrNot'] = df['LeaveOrNot'].astype('object')
```
