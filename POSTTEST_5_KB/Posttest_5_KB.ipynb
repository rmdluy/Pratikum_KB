```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dp = pd.read_csv("players.csv")
```

```python
dp.head()
```

# Data Spliting

```python
X = dp[['Goals']]
Y = dp['Matches']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.7,random_state=0)
```

```python
print("Dimensi X_train : ", X_train.shape)
print("Dimensi X_test  : ", X_test.shape)
print("Dimensi y_train : ", Y_train.shape)
print("Dimensi y_test  : ", Y_test.shape)
```

# Linear Regresion

```python
from sklearn.linear_model import LinearRegression
```

```python
modelLing = LinearRegression()
modelLing.fit(X_train, Y_train)
```

```python
prediction = modelLing.predict(X_test)
```

```python
plt.scatter(X_test, Y_test, color="Green", label="Nilai Aktual")
plt.plot(X_test, prediction, color="Blue", label="Nilai Prediksi")

plt.legend()
plt.grid(True)
plt.show()
```
